```python
import cv2
import numpy as np
from skimage.restoration import denoise_tv_chambolle, unsupervised_wiener
from matplotlib import pyplot as plt
```

```python
# 1. Load the image
image_path = "C:/Users/sofin/Downloads/LoLi-Phone_release/LoLi-Phone_release/all_flame/test_flame/Huawei Mate 20 Pro/VID_20210209_185634/0338.jpeg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# 5. Visualize the results
plt.figure(figsize=(12, 8))
plt.subplot(1, 3, 1)
# plt.title("Original Image")
plt.imshow(image)
plt.axis("off")
```

```python
# 2. Preprocess: Enhance brightness and contrast
alpha = 1.5  # Contrast control
beta = 30    # Brightness control
enhanced_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# 3. Denoise: Total Variation (TV) Denoising
tv_denoised_image = denoise_tv_chambolle(enhanced_image, weight=0.1, channel_axis=-1)
```

```python
plt.subplot(1, 3, 2)
# plt.title("Enhanced & Denoised")
plt.imshow(tv_denoised_image)
plt.axis("off")
```
